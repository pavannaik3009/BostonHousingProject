```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston #scikitlearn dataset of boston house pricing
```

```python
boston_data = load_boston()
```

```python
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
```

```python
y = boston_data.target #median house pricing
```

```python
y
```

## STATS Models

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_constant = sm.add_constant(X)
```

```python
pd.DataFrame(X_constant)
```

```python
model = sm.OLS(y, X_constant)
```

```python
lr = model.fit()
```

```python
lr.summary()
```

#### STATSFORMULA.API

```python
from_lr = smf.ols(formula = 'y ~ CRIM + ZN +INDUS + CHAS + NOX + RM', 
              data = df)
mlr = from_lr.fit()
```

## Correlation Matrix

```python
pd.options.display.float_format = '{:,.4f}'.format
corr_matrix = df.corr()
corr_matrix
```

```python
corr_matrix[np.abs(corr_matrix < 0.6)] = 0 #Mask the correlation matrix to highlight those > 0.6 corr. coef.
corr_matrix
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu') #Heatmap
plt.show()
```

#### Detecting Collinearity with Eigenvectors 

```python
eigenvalues, eigenvectors = np.linalg.eig(df.corr())
```

```python
pd.Series(eigenvalues).sort_values
```

```python
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending=False)
```

#### Standardise Variable to Identify key features (factor on the model)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns=['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
standard_coef_lin_reg = make_pipeline(scaler, model)
```

```python
standard_coef_lin_reg.fit(X, y)
result = pd.DataFrame(list(zip(standard_coef_lin_reg.steps[1][1].coef_, df.columns)), 
                     columns = ['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by='coefficient', ascending=False)
```

```python
from sklearn.metrics import r2_score
lr = smf.ols(formula = 'y ~ CRIM + ZN +INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT + AGE', 
              data = df)
lrf = lr.fit()
r2_score(y, lrf.predict(df))
```
